## Import and process Wildand Smoke Exposure data

```python
data_smoke = pd.read_csv('HMS_USpopBG_2010_2019.csv',usecols=['date','STATEFP','POPULATION','light','medium','heavy'])

# calculating impact of smoke on population
data_smoke['light_impact'] = data_smoke[(data_smoke['heavy'].isnull() == True) & (data_smoke['medium'].isnull() == True)].iloc[:,3] * data_smoke[(data_smoke['heavy'].isnull() == True) & (data_smoke['medium'].isnull() == True)].iloc[:,2]
data_smoke['medium_impact'] = data_smoke[(data_smoke['heavy'].isnull() == True) & (data_smoke['light'].isnull() == False)].iloc[:,4] * data_smoke[(data_smoke['heavy'].isnull() == True) & (data_smoke['light'].isnull() == False)].iloc[:,2]*10
data_smoke['heavy_impact'] = data_smoke[(data_smoke['light'].isnull() == False) & (data_smoke['medium'].isnull() == False)].iloc[:,5] * data_smoke[(data_smoke['light'].isnull() == False) & (data_smoke['medium'].isnull() == False)].iloc[:,2]*100
data_smoke = data_smoke.drop(columns=['POPULATION','light','medium','heavy'])
data_smoke = data_smoke.fillna(0)
data_smoke['impact'] = data_smoke['light_impact'] + data_smoke['medium_impact'] +data_smoke['heavy_impact']
data_smoke = data_smoke.drop(columns=['light_impact','medium_impact','heavy_impact'])

# extracting month, year from date
data_smoke['MONTH'] = data_smoke['date'].apply(str).str.slice(4,6).apply(int)
data_smoke['YEAR'] = data_smoke['date'].apply(str).str.slice(0,4).apply(int)
data_smoke = data_smoke.drop(columns=['date'])

data_smoke = data_smoke.rename(columns={'STATEFP':'USASTATE'})

# aggregating on state, year, month 
data_smoke2 = data_smoke.groupby(['USASTATE','YEAR','MONTH']).sum()

data_smoke2
```

## Union between data sets

```python
# union between trade and smoke data_smoke on state, year, month
final_data = data_trade.join(data_smoke2,on=['USASTATE','YEAR','MONTH'],how='outer')
```

```python
# ignoring USASTATE == 72 and dropping data items before 6/1/2010
final_data = final_data[(final_data['USASTATE'] != 72)][2032:]
```

```python
# adding sequence number for conversion to time-series
final_data['month_seq'] = ((final_data['YEAR']-2010)*12) + final_data['MONTH'] -6
```

```python
# rearranging columns for target as last column and replacing NaN with 0
final_data = final_data[['TRDTYPE','USASTATE','YEAR','MONTH','month_seq','COUNTRY','DISAGMOT','impact','VALUE']]
final_data = final_data.fillna(0)

final_data
```
